*CHECKING FOR THE MOST FREQUENT WORDS USING NLTK LIBRARY*

```python
import nltk
from nltk import FreqDist

def freq_words(x, terms = 30):
  all_words = ' '.join([text for text in x])
  all_words = all_words.split()

  fdist = FreqDist(all_words)
  words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())}) #like bow concept

  # selecting top 30 most frequent words
  d = words_df.nlargest(columns="count", n = terms) #nlargest sorts in descending order
  plt.figure(figsize=(20,5))
  ax = sns.barplot(data=d, x= "word", y = "count")
  ax.set(ylabel = 'Count')
  plt.show()
```

# *TOPIC MODELLING USING GENSIM - LDA APPROACH*

```python
data = grocery_cluster_data.copy()
data.shape
```

*IMPORTING REQUIRED LIBRARIES*

```python
!pip install -U gensim
import gensim
from gensim import corpora
!pip install pyLDAvis
import pyLDAvis
import pyLDAvis.gensim
```

```python
#simple preprocessing using gensim library
def preprocess(text):
   result = []
   for token in gensim.utils.simple_preprocess(text):
      if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 2:
         result.append(token)
   return result

data['norm_reviews'] = data['norm_reviews'].map(preprocess)
```

```python
#dictionary creation id2 word
dictionary = corpora.Dictionary(data['norm_reviews'])
#creating corpus term doucment frequency
doc_term_matrix = [dictionary.doc2bow(rev) for rev in data['norm_reviews']]#corpus
#contains bow # text to numer conversion is done here
#The doc2bow() function converts dictionary into a bag-of-words. The result, corpus, is a list of vectors equal to the number of documents.
#In each document vector is a series of tuples.
#building the LDA model
LDA = gensim.models.ldamodel.LdaModel
lda_model = LDA(corpus=doc_term_matrix, id2word=dictionary, num_topics=7, random_state=100,passes=25,chunksize=1000)
#passes: optional. The number of laps the model will take through corpus.
#The greater the number of passes, the more accurate the model will be. A lot of passes can be slow on a very large corpus.
#chunksize (int, optional) – Number of documents to be used in each training chunk.
```

```python
#Examining the results
#Our LDA model is now stored as ldamodel. Review our topics with the print_topic and print_topics methods:
lda_model.print_topics()
```

```python
# Compute Perplexity
print('\nPerplexity: ', lda_model.log_perplexity(doc_term_matrix))  # a measure of how good the model is. lower the better. #used for language model evaluation

# Compute Coherence Score Topic Coherence measures score a single topic by measuring the degree of semantic similarity between high scoring words in the topic.
from gensim.models import CoherenceModel
coherence_model_lda = CoherenceModel(model=lda_model, texts= data['norm_reviews'], dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```

```python
# Visualize the topics
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, doc_term_matrix, dictionary)
vis
```

*OPTIMAL K VALUE FOR LDA MODEL*

```python
def compute_coherence_values(dictionary, corpus, text_data, limit, start=2, step=3):
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = LDA(corpus=doc_term_matrix, id2word=dictionary, num_topics=num_topics,random_state=100,passes=25,chunksize=1000)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=text_data, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())
    return model_list, coherence_values
```

```python
model_list, coherence_values = compute_coherence_values(dictionary=dictionary, corpus=doc_term_matrix, text_data=data['norm_reviews'], start=2, limit=15, step=3)
# Ideal K value graph
limit=15; start=2; step=3;
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```

```python
#coherence scores
for t, cv in zip(x, coherence_values):
    print("Num Topics =", t, " has Coherence Value of", cv)
```

# *DOMINANT TOPIC IN EACH SENTENCE*

```python
def dominant_topics(ldamodel,corpus,texts):
  # Init output
  topics_data = pd.DataFrame()

  # Get main topic in each document
  for i, row in enumerate(ldamodel[corpus]):
      row = sorted(row, key=lambda x: (x[1]), reverse=True)
      # Get the Dominant topic, Perc Contribution and Keywords for each document
      for j, (topic_num, prop_topic) in enumerate(row):
          if j == 0:  # => dominant topic
              wp = ldamodel.show_topic(topic_num)
              topic_keywords = ", ".join([word for word, prop in wp])
              topics_data = topics_data.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
          else:
              break
  topics_data.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

  # Add original text to the end of the output
  contents = pd.Series(texts)
  topics_data = pd.concat([topics_data, contents], axis=1)
  return(topics_data)
```

```python
#optimal model
optimal_model = model_list[3] #k=3
```

```python
topic_keywords = dominant_topics(ldamodel = optimal_model, corpus = doc_term_matrix, texts = data['norm_reviews'])

# Format
topics_grocery_info = topic_keywords.reset_index()
topics_grocery_info.columns = ['Document_No', 'Dominant_Topic', 'Topic_Perc_Contrib', 'Keywords', 'Text']
topics_grocery_info.head(10)
```

*FINDING MOST REPRESENTATIVE TEXT DOCUMENT FOR EACH TOPIC*

```python
# Group top 5 sentences under each topic
topic_info = pd.DataFrame()

topic_grpd = topics_grocery_info.groupby('Dominant_Topic')

for i, grp in topic_grpd:
  topic_info = pd.concat([topic_info, grp.sort_values(['Topic_Perc_Contrib'], ascending=[0]).head(1)],axis=0)

# Reset Index    
topic_info.reset_index(drop=True, inplace=True)

# Format
topic_info.columns = ['Doc_no','Topic_Num', "Topic_Perc_Contrib", "Keywords", "Text"]

# Show
topic_info
```

# *Topic distribution across documents*

```python
# Number of Documents for Each Topic
topic_counts = topics_grocery_info['Dominant_Topic'].value_counts()

# Percentage of Documents for Each Topic
topic_contribution = round(topic_counts/topic_counts.sum()*100,3)

# Topic Number and Keywords
topic_num_keywords = topic_info[['Topic_Num', 'Keywords']]

# Concatenate Column wise
topic_dominancy = pd.concat([topic_num_keywords, topic_counts, topic_contribution], axis=1)

# Change Column names
topic_dominancy.columns = ['Dominant_Topic', 'Topic_Keywords', 'Num_Documents', 'Percentage_of_total_Documents']

# Show
topic_dominancy
```
